# How Are We Calculating It?

We are using **Cohens d**, which measures how different two means are in terms of standard deviations.

## Formula:
\[
d = \frac{\text{Mean(TS_True)} - \text{Mean(TS_False)}}{\text{Pooled Standard Deviation}}
\]

### Where:
- **Numerator**: The difference in means (how much higher TS_True is than TS_False).
- **Denominator**: The pooled standard deviation, which accounts for the spread of values in both TS_True and TS_False.

The larger the effect size, the greater the difference between TS_True and TS_False in terms of standard deviations.

---

## Interpreting Cohens d

| Effect Size (d) | Interpretation                  |
|-----------------|---------------------------------|
| 0.0 - 0.2       | Small effect (minimal impact)   |
| 0.2 - 0.5       | Medium effect (moderate impact) |
| 0.5 - 0.8       | Large effect (strong impact)    |
| > 0.8           | Very large effect (significant impact) |

### For example:
- If **Effect_Size = 0.7**, it means TS_True increases shooting percentage by **0.7 standard deviations**, which is a moderate-to-large effect.
- If **Effect_Size = 1.5**, it suggests a **very strong impact** of TS on shooting percentage.

---

## Why Is This Useful?

### **Better Than Just Looking at Percent Increase**
- A high percent increase (e.g., 200%) can be misleading if the baseline is very small.
- Effect size normalizes the difference by considering how much shooting percentages vary overall.

### **Makes It Easy to Compare Different Chance Types**
- Even if one chance type has a small absolute increase, if it has a high effect size, it means TS has a significant impact on that type of shot.

```python
import numpy as np

# Ensure numeric columns are treated as floats
numeric_cols = [
    'Shooting Percent_all', 
    'Historical % (All)', 
    'Shooting Percent_TS_True', 
    'Historical % (TS_True)', 
    'Shooting Percent_TS_False', 
    'Historical % (TS_False)'
]
shooting_historical_df[numeric_cols] = shooting_historical_df[numeric_cols].apply(pd.to_numeric, errors='coerce')

# Function to calculate Cohen's d for each row
def cohens_d_row(current, historical):
    if current == 0 and historical == 0:
        return 0  # Avoid division by zero
    pooled_std = np.sqrt((current**2 + historical**2) / 2)
    return (current - historical) / pooled_std

# Calculate Cohen's d for each comparison
shooting_historical_df['Cohen_d_All'] = shooting_historical_df.apply(
    lambda row: cohens_d_row(row['Shooting Percent_all'], row['Historical % (All)']), axis=1
)
shooting_historical_df['Cohen_d_TS_True'] = shooting_historical_df.apply(
    lambda row: cohens_d_row(row['Shooting Percent_TS_True'], row['Historical % (TS_True)']), axis=1
)
shooting_historical_df['Cohen_d_TS_False'] = shooting_historical_df.apply(
    lambda row: cohens_d_row(row['Shooting Percent_TS_False'], row['Historical % (TS_False)']), axis=1
)

# Function to categorize Cohen's d
def categorize_cohens_d(value):
    if abs(value) < 0.2:
        return "Small"
    elif abs(value) < 0.5:
        return "Medium"
    elif abs(value) < 0.8:
        return "Large"
    else:
        return "Very Large"

# Apply categorization to Cohen's d columns
shooting_historical_df['Cohen_d_All_Category'] = shooting_historical_df['Cohen_d_All'].apply(categorize_cohens_d)
shooting_historical_df['Cohen_d_TS_True_Category'] = shooting_historical_df['Cohen_d_TS_True'].apply(categorize_cohens_d)
shooting_historical_df['Cohen_d_TS_False_Category'] = shooting_historical_df['Cohen_d_TS_False'].apply(categorize_cohens_d)

# Remove the Effect_Size_Category column
# shooting_historical_df = shooting_historical_df.drop(columns=['Effect_Size_Category'])

# Remove unnecessary trailing zeros from numeric columns
def remove_trailing_zeros(value):
    if isinstance(value, (float, np.floating)):
        return f'{value:.6f}'.rstrip('0').rstrip('.') if '.' in f'{value:.6f}' else f'{value:.6f}'
    elif isinstance(value, str):
        return value.rstrip('0').rstrip('.') if '.' in value else value
    return value

# Apply the function to clean up trailing zeros
shooting_historical_df = shooting_historical_df.applymap(remove_trailing_zeros)

# Split into three DataFrames
df_all = shooting_historical_df[['Chance Type', 'Shooting Percent_all', 'Historical % (All)', 'Cohen_d_All', 'Cohen_d_All_Category']]
df_ts_true = shooting_historical_df[['Chance Type', 'Shooting Percent_TS_True', 'Historical % (TS_True)', 'Cohen_d_TS_True', 'Cohen_d_TS_True_Category']]
df_ts_false = shooting_historical_df[['Chance Type', 'Shooting Percent_TS_False', 'Historical % (TS_False)', 'Cohen_d_TS_False', 'Cohen_d_TS_False_Category']]

# Apply highlights to critical shot types and specific columns
def highlight_styles(row):
    styles = []
    for col in row.index:
        if row['Chance Type'] in critical_shot_types:
            styles.append('background-color: lightyellow')
        elif col == 'Shooting Percent_TS_True':
            styles.append('background-color: #d4edda')  # light green
        elif col == 'Shooting Percent_TS_False':
            styles.append('background-color: #f8d7da')  # light red
        elif col == 'Historical % (TS_True)':
            styles.append('background-color: #d4edda')  # light green
        elif col == 'Historical % (TS_False)':
            styles.append('background-color: #f8d7da')  # light red
        else:
            styles.append('')
    return styles

# Apply the styling to each DataFrame
styled_df_all = df_all.style.apply(highlight_styles, axis=1)
styled_df_ts_true = df_ts_true.style.apply(highlight_styles, axis=1)
styled_df_ts_false = df_ts_false.style.apply(highlight_styles, axis=1)

# Display the updated dataframes with metrics
print("Shooting Percent Comparison: All Shots")
display(styled_df_all)

print("\nShooting Percent Comparison: TS_True Shots")
display(styled_df_ts_true)

print("\nShooting Percent Comparison: TS_False Shots")
display(styled_df_ts_false)
```
