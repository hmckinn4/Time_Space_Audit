```python
import pandas as pd

# Load the uploaded files
true_df = pd.read_excel('TS_True_All_Goals_Chances.xlsx')
false_df = pd.read_excel('TS_False_All_Goals_Chances.xlsx')
all_df = pd.read_excel('All_Goals_Chances.xlsx')
```

```python
# Define critical shot types
critical_shot_types = [
    "Defensive Screen", "High Slot Line Pass", "Layered Screen", "Mid % Broken Play", "Low Slot Line Pass",
    "Breakaway", "Half-Ice Breakaway", "CS Odd Man Rush Slot",
    "Behind the Net To Slot One Timer", "Slot Line Double", "Deflection On Ice", "CS Broken Play",
    "One Timer Outside Slot Area", "Partial Breakaway"
]
```

```python
print("TS_False_All_Goals_Chances Data")
false_df = false_df.round(2) 
false_df
```

```python
# Rename original dataframes
all_df_historical = all_df.copy()
true_df_historical = true_df.copy()
false_df_historical = false_df.copy()

# Compute Change_in_Shooting% for each dataframe
all_df_historical["Change_in_Shooting%"] = all_df_historical["Shooting Percent"] - all_df_historical["Historical %"]
true_df_historical["Change_in_Shooting%"] = true_df_historical["Shooting Percent"] - true_df_historical["Historical %"]
false_df_historical["Change_in_Shooting%"] = false_df_historical["Shooting Percent"] - false_df_historical["Historical %"]

# Define columns to drop
columns_to_drop = ["All Chances", "Goals", "Expected Goals", "Goals Diff Toggle"]

# Drop columns, remove NaNs, and sort by Change_in_Shooting%
all_df_historical = (
    all_df_historical
    .drop(columns=columns_to_drop, errors='ignore')
    .dropna()
    .query("`Chance Type` != 'Goalie Error'")  # Remove rows where "Chance Type" is "Goalie Error"
    .sort_values(by="Change_in_Shooting%", ascending=False)
)

true_df_historical = (
    true_df_historical
    .drop(columns=columns_to_drop, errors='ignore')
    .dropna()
    .query("`Chance Type` != 'Goalie Error'")
    .sort_values(by="Change_in_Shooting%", ascending=False)
)

false_df_historical = (
    false_df_historical
    .drop(columns=columns_to_drop, errors='ignore')
    .dropna()
    .query("`Chance Type` != 'Goalie Error'")
    .sort_values(by="Change_in_Shooting%", ascending=False)
)
```

```python
# Cell 7: Visualizing Change_in_Shooting% as a Bar Chart with Categorized Colors

import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.patches import Patch

# Define color categories based on Change_in_Shooting% values
colors_map = {
    "Small": "#f8d7da",  # Light Red
    "Medium": "#fff3cd",  # Light Yellow
    "Large": "#d4edda"  # Light Green
}

def categorize_effect_size(value):
    """Assign a color category based on Change_in_Shooting% value."""
    if value <= 0.08:
        return "Small"
    elif 0.08 < value <= 0.19:
        return "Medium"
    else:  # value > 0.19
        return "Large"

# Function to create a categorized bar chart for a given dataframe
def plot_change_in_shooting(df, title):
    # Convert Change_in_Shooting% to numeric (if needed)
    df['Change_in_Shooting%'] = pd.to_numeric(df['Change_in_Shooting%'], errors='coerce')

    # Sort dataframe by Change_in_Shooting% in descending order
    df = df.sort_values(by='Change_in_Shooting%', ascending=False)

    # Apply categorization and colors
    df['Effect_Size_Category'] = df['Change_in_Shooting%'].apply(categorize_effect_size)
    colors = df['Effect_Size_Category'].map(colors_map)

    # Plot bar chart
    plt.figure(figsize=(12, 6))
    bars = plt.barh(df['Chance Type'], df['Change_in_Shooting%'], color=colors)

    # Highlight critical shot types in red
    for i, bar in enumerate(bars):
        chance_type = df['Chance Type'].iloc[i]
        if chance_type in critical_shot_types:
            plt.gca().get_yticklabels()[i].set_color("red")
            plt.gca().get_yticklabels()[i].set_fontweight("bold")  # Make text bold

    # Labels and Title
    plt.xlabel("Change in Shooting % (Shooting Percent - Historical %)")
    plt.ylabel("Chance Type")
    plt.title(title)

    # Add a legend
    legend_elements = [
        Patch(facecolor=colors_map["Small"], label="Small Change ( 0.08)"),
        Patch(facecolor=colors_map["Medium"], label="Medium Change (0.08 - 0.19)"),
        Patch(facecolor=colors_map["Large"], label="Large Change (> 0.19)")
    ]
    plt.legend(handles=legend_elements, loc="lower right")

    # Invert y-axis to show highest increases at the top
    plt.gca().invert_yaxis()

    # Show plot
    plt.show()

# Generate bar charts for each dataset
plot_change_in_shooting(all_df_historical, "Change in Shooting % Across All Chances")
plot_change_in_shooting(true_df_historical, "Change in Shooting % When TS is True")
plot_change_in_shooting(false_df_historical, "Change in Shooting % When TS is False")
```

```python
import pandas as pd

# Ensure numeric columns are floats
for df in [all_df_historical, true_df_historical, false_df_historical]:
    df[["Shooting Percent", "Historical %", "Change_in_Shooting%"]] = df[["Shooting Percent", "Historical %", "Change_in_Shooting%"]].apply(pd.to_numeric, errors="coerce")

# Define color categories based on individual values
colors_map = {
    "Small": "#f8d7da",  # Light Red
    "Medium": "#fff3cd",  # Light Yellow
    "Large": "#d4edda"  # Light Green
}

# Define critical shot types
critical_shot_types = [
    "Defensive Screen", "High Slot Line Pass", "Layered Screen", "Mid % Broken Play", "Low Slot Line Pass",
    "Breakaway", "Half-Ice Breakaway", "CS Odd Man Rush Slot",
    "Behind the Net To Slot One Timer", "Slot Line Double", "Deflection On Ice", "CS Broken Play",
    "One Timer Outside Slot Area", "Partial Breakaway"
]

def categorize_effect_size(value):
    """Assign a color category based on value."""
    if value <= 0.08:
        return "Small"  # Light Red
    elif 0.08 < value <= 0.19:
        return "Medium"  # Light Yellow
    else:  # value > 0.19
        return "Large"  # Light Green

# Function to highlight numeric cells
def highlight_individual_cells(data):
    styles = pd.DataFrame("", index=data.index, columns=data.columns)  # Default empty styles
    for col in ["Shooting Percent", "Historical %", "Change_in_Shooting%"]:  # Apply to these columns
        styles[col] = data[col].apply(lambda x: f'background-color: {colors_map[categorize_effect_size(float(x))]}')
    return styles

# Function to bold and color critical shot types
def highlight_critical_shots(data):
    styles = [""] * len(data)  # Default empty styles
    for i, shot in enumerate(data):
        if shot in critical_shot_types:
            styles[i] = "font-weight: bold; color: red"
    return styles

# Define table titles
table_titles = {
    "All Data": "All Chances",
    "TS True": "Time and Space True",
    "TS False": "Time and Space False"
}

# Apply sorting, formatting, and display each dataframe with titles
for df_name, df in {"All Data": all_df_historical, "TS True": true_df_historical, "TS False": false_df_historical}.items():
    # Ensure Change_in_Shooting% is float for correct sorting
    df["Change_in_Shooting%"] = pd.to_numeric(df["Change_in_Shooting%"], errors="coerce")

    # Split the dataframe into positive (including zero) and negative values
    positive_df = df[df["Change_in_Shooting%"] >= 0]
    negative_df = df[df["Change_in_Shooting%"] < 0]

    # Sort positive values in descending order
    positive_df = positive_df.sort_values(by="Change_in_Shooting%", ascending=False)

    # Sort negative values in ascending order
    negative_df = negative_df.sort_values(by="Change_in_Shooting%", ascending=False)

    # Concatenate the sorted positive and negative dataframes
    sorted_df = pd.concat([positive_df, negative_df])

    # Format for display (removing trailing zeros but keeping values numeric)
    sorted_df[["Shooting Percent", "Historical %", "Change_in_Shooting%"]] = sorted_df[
        ["Shooting Percent", "Historical %", "Change_in_Shooting%"]
    ].applymap(lambda x: f"{x:.6g}" if isinstance(x, float) else x)

    # Apply the styles
    styled_df = sorted_df.style.apply(highlight_individual_cells, axis=None)  # Apply color coding to numeric cells
    styled_df = styled_df.apply(highlight_critical_shots, subset=["Chance Type"], axis=0)  # Bold red critical shots
    
    # Print table title
    print(f"\n### {table_titles[df_name]}")  # Display title before each table
    
    # Display dataframe
    display(styled_df)
```
