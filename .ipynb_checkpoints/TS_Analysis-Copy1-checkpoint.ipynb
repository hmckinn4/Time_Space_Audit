# Time and Space Chances & Goals Audit 02/18/25 for 24-25 NHL Season

```python
import pandas as pd

# Load the uploaded files
true_df = pd.read_excel('TS_True_All_Goals_Chances.xlsx')
false_df = pd.read_excel('TS_False_All_Goals_Chances.xlsx')
all_df = pd.read_excel('All_Goals_Chances.xlsx')
```

```python
# Define critical shot types
critical_shot_types = [
    "Defensive Screen", "High Slot Line Pass", "Layered Screen", "Mid % BP", "Low SL Pass",
    "Breakaway", "Half Ice Breakaway (Half ice partial for the trackers)", "CS Odd Man Rush Slot",
    "BTN To Slot One Timer", "Sl Double", "Deflection On Ice", "CS Broken Play",
    "One Timer Outside Slot Area", "Partial Breakaway"
]
```

# All Data Displayed

```python
# Cell 2: Display All_Goals_Chances data (rounded) with highlights
print("All_Goals_Chances Data")

# Round results to 2 decimal places, remove trailing zeros
all_df = all_df.applymap(lambda x: f'{x:.2f}'.rstrip('0').rstrip('.') if isinstance(x, float) else x)

# Shorten long tuple-like strings for Chance Type and simplify display
all_df['Chance Type'] = all_df['Chance Type'].astype(str).str.replace(r"[()']", '', regex=True).str.replace(',', ' -').str.slice(0, 50)

# Apply highlights to critical shot types using style.apply
def highlight_styles(row):
    return ['background-color: lightyellow' if row['Chance Type'] in critical_shot_types else '' for _ in row]

all_df_styled = all_df.style.apply(highlight_styles, axis=1)

# Display all shot types, highlighting only critical ones
all_df = all_df_styled

display(all_df)
```

# Time and Space True Data Displayed

```python
# Display the loaded tables for TS_True_All_Goals_Chances with highlights
print("TS_True_All_Goals_Chances Data")

# Round results to 2 decimal places, remove trailing zeros
true_df = true_df.applymap(lambda x: f'{x:.2f}'.rstrip('0').rstrip('.') if isinstance(x, float) else x)

# Shorten long tuple-like strings for Chance Type and simplify display
true_df['Chance Type'] = true_df['Chance Type'].astype(str).str.replace(r"[()']", '', regex=True).str.replace(',', ' -').str.slice(0, 50)

# Apply highlights to critical shot types using style.apply
def highlight_styles(row):
    return ['background-color: lightyellow' if row['Chance Type'] in critical_shot_types else '' for _ in row]

true_df_styled = true_df.style.apply(highlight_styles, axis=1)

# Display all shot types, highlighting only critical ones
true_df = true_df_styled
display(true_df)
```

# Time and Space False Data Displayed

```python
# Display the loaded tables for TS_False_All_Goals_Chances with highlights
print("TS_False_All_Goals_Chances Data")

# Round results to 2 decimal places, remove trailing zeros
false_df = false_df.applymap(lambda x: f'{x:.2f}'.rstrip('0').rstrip('.') if isinstance(x, float) else x)

# Shorten long tuple-like strings for Chance Type and simplify display
false_df['Chance Type'] = false_df['Chance Type'].astype(str).str.replace(r"[()']", '', regex=True).str.replace(',', ' -').str.slice(0, 50)

# Apply highlights to critical shot types using style.apply
def highlight_styles(row):
    return ['background-color: lightyellow' if row['Chance Type'] in critical_shot_types else '' for _ in row]

false_df_styled = false_df.style.apply(highlight_styles, axis=1)

# Display all shot types, highlighting only critical ones
false_df = false_df_styled
display(false_df)
```

```python
all_df = all_df.round(2) 
true_df = true_df.round(2)
false_df = false_df.round(2) 
```

# Shooting Percentage Comparison

```python
# Cell 5: Create shooting percent comparison table
shooting_comparison_df = pd.merge(
    pd.merge(
        all_df[['Chance Type', 'Shooting Percent']],
        true_df[['Chance Type', 'Shooting Percent']],
        on='Chance Type',
        suffixes=('_all', '_TS_True')
    ),
    false_df[['Chance Type', 'Shooting Percent']],
    on='Chance Type'
)
shooting_comparison_df.rename(columns={'Shooting Percent': 'Shooting Percent_TS_False'}, inplace=True)

# Calculate differences
shooting_comparison_df['TS_True_minus_all'] = (
    shooting_comparison_df['Shooting Percent_TS_True'] - shooting_comparison_df['Shooting Percent_all']
)
shooting_comparison_df['TS_True_minus_TS_False'] = (
    shooting_comparison_df['Shooting Percent_TS_True'] - shooting_comparison_df['Shooting Percent_TS_False']
)
shooting_comparison_df['all_minus_TS_False'] = (
    shooting_comparison_df['Shooting Percent_all'] - shooting_comparison_df['Shooting Percent_TS_False']
)

# Round results to 3 decimal places and remove trailing zeros
shooting_comparison_df = shooting_comparison_df.applymap(lambda x: f'{x:.3f}'.rstrip('0').rstrip('.') if isinstance(x, float) else x)

# Highlight critical shot types and specific columns
def highlight_styles(row):
    styles = []
    for col in shooting_comparison_df.columns:
        if row['Chance Type'] in critical_shot_types:
            styles.append('background-color: lightyellow')
        elif col == 'Shooting Percent_TS_True':
            styles.append('background-color: #d4edda')  # light green
        elif col == 'Shooting Percent_TS_False':
            styles.append('background-color: #f8d7da')  # light red
        else:
            styles.append('')
    return styles

styled_df = shooting_comparison_df.style.apply(highlight_styles, axis=1)

# Display the comparison results with highlights
print("Shooting Percent Comparison (Critical Shot Types Highlighted and Column Colors Applied)")
display(styled_df)
```
